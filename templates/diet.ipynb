```python
import random # You'll need this to pick random recommendations later

def recommend_food(user_input):
    # --- 1. Extract Inputs (Correctly indented) ---
    age = user_input.get('Age')
    gender = user_input.get('Gender')
    height = user_input.get('Height')
    weight = user_input.get('Weight')
    # NOTE: Your form returns BMI, Blood Pressure, Cholesterol, and Sugar as strings (e.g., 'High', 'Overweight').
    # We use them as strings here.
    bmi = user_input.get('BMI')
    chronicDisease = user_input.get('Chronic_Disease') # Use chronicDisease for consistency
    bp = user_input.get('Blood_Pressure')
    cholesterol = user_input.get('Cholesterol')
    sugar = user_input.get('Sugar')
    sleep = user_input.get('Sleep_Hours')
    alcohol = user_input.get('Alcohol')
    smoking = user_input.get('Smoking')
    stress_eating = user_input.get('Stress_Eating')
    diet = user_input.get('Dietary_Habits')
    sodium_sensitive = user_input.get('Sodium_Sensitivity')
    sugar_sensitive = user_input.get('Sugar_Sensitivity')

    # --- 2. Expanded and Classified Food Pool ---
    # NOTE: These lists are defined and kept INSIDE the function.
    
    # --- Breakfast Options ---
    # Start by making copies of the base lists so we can filter them safely.
    breakfast_foods = [
        'Oats porridge with berries and nuts', 'Vegetable upma', 
        'Idli with sambar and coconut chutney', 'Ragi dosa',
        'Scrambled egg whites with spinach', 'Fruit salad with Greek yogurt (low-fat)',
        'Whole wheat toast with avocado and tomato', 'Barley porridge with cinnamon',
        'Moong dal chilla (pancake) with mint chutney', 'Smoothie (Spinach, banana, chia seeds)'
    ]

    # --- Lunch Options ---
    lunch_foods = [
        'Brown rice with dal (lentil soup) and mixed vegetable curry',
        'Chapati (whole wheat) with spinach curry (Palak Paneer/Tofu)',
        'Quinoa salad with roasted vegetables and lemon vinaigrette', 'Kidney beans (Rajma) with brown rice',
        'Lentil soup (Dal) with sorghum (Jowar) bhakri', 'Grilled tofu/paneer tikka salad',
        'Whole wheat pasta with lean ground turkey/vegetables', 'Vegetable pulav with raita (low-fat yogurt)',
        'Mackerel or Salmon curry with millet'
    ]

    # --- Dinner Options ---
    dinner_foods = [
        'Khichdi with turmeric and steamed green beans', 'Rasam with millets (thin soup with spices)',
        'Grilled chicken breast/tofu with sweet potato mash and steamed broccoli',
        'Chickpea curry (Chana Masala) with whole wheat roti', 'Baked salmon with roasted asparagus and brown rice',
        'Mixed vegetable stew (Sambar) with Ragi mudde (balls)', 'Lentil soup (Masoor Dal) with a side of cucumber salad',
        'SautÃ©ed vegetables (broccoli, bell peppers) with quinoa', 'Turkey/Tofu chili (bean-based)'
    ]
    
    # Create a list of the meal lists to easily apply the filters to all three
    meal_lists = [breakfast_foods, lunch_foods, dinner_foods]
    
    # --- 3. Filters (Apply to all three lists within a loop) ---
    
    filtered_meals = []
    
    for meal_foods in meal_lists:
        
        # --- FILTER 1: High Cholesterol / Fats (Heart Health) ---
        if cholesterol == 'High':
            meal_foods = [f for f in meal_foods if 'cheese' not in f.lower() and 'fried' not in f.lower() and 'butter' not in f.lower() and 'creamy' not in f.lower()]

        # --- FILTER 2: Hypertension (High BP) and Sodium Sensitivity ---
        if chronicDisease == 'Hypertension' or sodium_sensitive == 'yes' or bp == 'High':
            # Strict low-sodium filter (removes salty condiments and preserved foods)
            meal_foods = [f for f in meal_foods if 'pickle' not in f.lower() and 'papad' not in f.lower() and 'sauce' not in f.lower() and 'processed' not in f.lower() and 'salt' not in f.lower()]

        # --- FILTER 3: Diabetes (High Sugar) and Sugar Sensitivity ---
        if chronicDisease == 'Diabetes' or sugar == 'High' or sugar_sensitive == 'yes':
            # Strict low-glycemic/low-sugar filter
            meal_foods = [f for f in meal_foods if 'sweet' not in f.lower() and 'juice' not in f.lower() and 'sugar' not in f.lower() and 'refined' not in f.lower()]

        # --- FILTER 4: Obesity (BMI) / Weight Management ---
        if chronicDisease == 'Obesity' or bmi in ['Overweight', 'Obese']:
            # Prioritize high fiber, low caloric density, and whole grains
            meal_foods = [f for f in meal_foods if 'rice' not in f.lower() or 'brown rice' in f.lower()]
            # Further filter high-calorie items
            meal_foods = [f for f in meal_foods if 'potatoes' not in f.lower() and 'white bread' not in f.lower() and 'oil' not in f.lower()]

        # --- FILTER 5: Asthma (General Respiratory Sensitivity) ---
        if chronicDisease == 'Asthma':
            # Recommended to avoid known inflammatory or phlegm-producing foods (dairy, cold items, fried)
            meal_foods = [f for f in meal_foods if 'fried' not in f.lower() and 'creamy' not in f.lower() and 'dairy' not in f.lower()]

        # --- FILTER 6: Dietary Habits (Vegetarian/Non-Vegetarian) ---
        if diet == 'Vegetarian':
            # Ensure no non-vegetarian protein sources remain
            meal_foods = [f for f in meal_foods if 'fish' not in f.lower() and 'egg' not in f.lower() and 'chicken' not in f.lower() and 'salmon' not in f.lower() and 'turkey' not in f.lower()]

        # --- FILTER 7: Stress Eating (Behavioral - Additions) ---
        if stress_eating == 'yes':
            meal_foods += ['Warm oats with banana', 'Chamomile tea', 'Herbal soup with lentils'] 
        
        # Append the filtered list to the result
        filtered_meals.append(meal_foods)
        
    # --- 4. Return Recommendations ---
    
    # Unpack the list of lists back into named variables for clarity
    filtered_breakfast = filtered_meals[0]
    filtered_lunch = filtered_meals[1]
    filtered_dinner = filtered_meals[2]
    
    # Pick a random recommendation from each (if the list isn't empty)
    recommendations = {
        'Breakfast': random.choice(filtered_breakfast) if filtered_breakfast else "No specific breakfast found based on filters.",
        'Lunch': random.choice(filtered_lunch) if filtered_lunch else "No specific lunch found based on filters.",
        'Dinner': random.choice(filtered_dinner) if filtered_dinner else "No specific dinner found based on filters."
    }
    
    # You can return this dictionary to the Flask endpoint
    return recommendations
```
